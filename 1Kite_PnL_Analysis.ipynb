```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from openpyxl import load_workbook
```

```python
df = pd.read_csv("orders (1).csv")
df
```

```python
# Only Show MIS and Complete data

f_df = df[(df["Product"] == "MIS") & (df["Status"] == "COMPLETE")].copy()
f_df
```

```python
f_df[["Executed_Qty", "Ordered_Qty"]] = f_df["Qty."].str.split("/", expand=True).astype(int)
f_df[["Executed_Qty", "Ordered_Qty"]]
```

```python
f_df
```

```python
# Brokerage 

f_df["Trade_Value"] = f_df["Executed_Qty"] * f_df["Avg. price"]
f_df["Brokerage"] = (f_df["Trade_Value"] * 0.0003).clip(upper=20)
f_df["Brokerage"] 
```

```python
f_df = f_df.loc[:,~f_df.columns.duplicated()].copy()
f_df
```

```python
f_df.info()
```

```python
#  Column ke rename karna hai 

f_df = f_df.rename(columns={
    'trade_value': 'Trade_Value'
    
})
f_df
```

```python
f_df["stt"] = np.where(
    f_df["Type"] == "SELL",
    f_df["Trade_Value"] * 0.00025,
    0
).round(2)

f_df["stt"]
```

```python
f_df = f_df.rename(columns={
    'stt': 'STT'
    
})
```

```python
f_df
```

```python
f_df["Transaction_Charges"] = (f_df["Trade_Value"] * 0.0000307).round(2)
f_df["Transaction_Charges"]
```

```python
f_df["SEBI"] = ((f_df["Trade_Value"] / 10000000) * 10).round(3)
f_df["SEBI"]
```

```python
f_df
```

```python
f_df["Stamp_Charges"] = np.where(f_df["Type"] == "BUY", 
                                f_df["Trade_Value"] * 0.00003,
                                0).round(2)
f_df["Stamp_Charges"]
```

```python
f_df
```

```python
f_df["GST"] = (
    (f_df["Brokerage"] + f_df["SEBI"] + f_df["Transaction_Charges"] * 0.18)
)
f_df["GST"]
```

```python
f_df
```

```python
f_df["Total_Charges"] = f_df["Brokerage"] + f_df["SEBI"] + f_df["Transaction_Charges"] + f_df["STT"] + f_df["Stamp_Charges"] + f_df["GST"]
f_df["Total_Charges"]
```

```python
f_df
```

```python
# 2) Stock wise and Type wise analysis with weighted Avg. price & calculated charges 
```

```python
analysis = f_df.groupby(["Instrument", "Type"]).agg(
    Qty=("Executed_Qty", "sum"),
    Turnover=("Trade_Value", "sum"),
    Brokerage=("Brokerage", "sum"),
    STT_CTT=("STT", "sum"),
    ETC=("Transaction_Charges", "sum"),
    SEBI=("SEBI", "sum"),
    GST=("GST", "sum"),
    Stamp_Charges=("Stamp_Charges", "sum"),
    Total_Charges=("Total_Charges", "sum")
).reset_index()

analysis
```

```python
Summary = f_df.groupby(["Instrument", "Type"]).agg(
    Turnover=("Trade_Value", "sum"),
    Total_Charges=("Total_Charges", "sum")
).reset_index()

Summary
```

```python
with pd.ExcelWriter("Kite_analysis_report.xlsx", engine="openpyxl") as writer:
    f_df.to_excel(writer, sheet_name="Raw_Data", index=False)
    analysis.to_excel(writer, sheet_name="Stock_Type_Analysis", index=False)
    Summary.to_excel(writer, sheet_name="Overall_Summary", index=False)

print("Excel file saved successfully!")
```

```python
summary = f_df.groupby(["Instrument", "Type"]).agg(
    Turnover=("Trade_Value", "sum"),
    Total_Charges=("Total_Charges", "sum")
).reset_index()

pivot = summary.pivot(
    index="Instrument",
    columns="Type",
    values="Turnover"
).fillna(0)

overall = f_df.groupby("Instrument")["Total_Charges"].sum().reset_index()

overall["Gross_PnL"] = (
    pivot.get("SELL", 0) - pivot.get("BUY", 0)
).values

overall["Net_PnL"] = overall["Gross_PnL"] - overall["Total_Charges"]

overall["%_Charges"] = (
    (overall["Total_Charges"] / overall["Gross_PnL"]) * 100
).round(2)

overall = overall[
    ["Instrument", "Gross_PnL", "Total_Charges", "Net_PnL", "%_Charges"]
]


total_row = pd.DataFrame({
    "Instrument": ["Total"],
    "Gross_PnL": [overall["Gross_PnL"].sum()],
    "Total_Charges": [overall["Total_Charges"].sum()],
    "Net_PnL": [overall["Net_PnL"].sum()],
    "%_Charges": [round((overall["Total_Charges"].sum() / overall["Gross_PnL"].sum()) * 100, 2)]
})

overall = pd.concat([overall, total_row], ignore_index=True)

overall
```

```python
with pd.ExcelWriter("Kite_analysis_report2.xlsx", engine="openpyxl") as writer:
    f_df.to_excel(writer, sheet_name="Raw_Data", index=False)
    analysis.to_excel(writer, sheet_name="Stock_Type_Analysis", index=False)
    overall.to_excel(writer, sheet_name="Overall_Summary", index=False)

print("Excel file saved successfully!")
```
